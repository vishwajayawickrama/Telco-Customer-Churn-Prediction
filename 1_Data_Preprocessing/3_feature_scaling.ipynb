# Feature Scalling

### 1. Install Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

### 2. Load Dataset

```python
df = pd.read_csv('../data/processed/TelcoCustomerChurnPredictionFeatureEncoded.csv')
df.head()
```

```python
numerical_variables = [
                        'MonthlyCharges', 'TotalCharges', 'tenure', "ServiceAdoptionScore", "AvgMonthlyChargesPerService", 
                      ]

for col in numerical_variables:
    standard_scalar = StandardScaler()
    df[col] = standard_scalar.fit_transform(df[col].values.reshape(df[col].shape[0], 1))

df.head()
```

```python
df.to_csv('../data/processed/TelcoCustomerChurnPredictionFeatureScaled.csv', index=False)
```
